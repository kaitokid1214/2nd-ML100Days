```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
#10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣; np.random.uniform(low, high, size)
matrix = np.random.uniform(-1.0, 1.0, size=[10, 10])
print(matrix)
plt.figure(figsize=(10, 10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True, vmin = -0.8, vmax = 0.8)
#cmap => colormap
#annot => 圖樣中色塊的數字是否顯示
#vmin, vmax =>小於 / 大於某個值後顯示像是hot zone一樣的表示
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
#1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
matrix = np.random.uniform(-1.0, 1.0, size=[nrow, ncol])
# print(matrix)
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)
#參考: https://seaborn.pydata.org/generated/seaborn.PairGrid.html
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
#1000 x 3, 數值分布為常態分佈的矩陣
#"常態分佈"參考: https://kknews.cc/zh-tw/tech/39v3vny.html
matrix = np.random.normal(size = [nrow, ncol])
# print(matrix)
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
